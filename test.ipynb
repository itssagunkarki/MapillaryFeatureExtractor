```python
import requests

def call_mapillary_api():
    url = "https://graph.mapillary.com/550092599700936"
    access_token = "MLY|7499350746781034|b63ef80c5ee40264229b0363235dd20d"
    params = {
        "access_token": access_token,
        "fields": "id,computed_geometry,detections.value"
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()  # Raise an exception for HTTP errors
        data = response.json()
        return data
    except requests.exceptions.RequestException as e:
        print("Error calling Mapillary API:", e)
        return None

# Call the function and print the result
api_data = call_mapillary_api()
if api_data:
    print(api_data)
```
